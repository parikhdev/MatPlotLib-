```python
import matplotlib.pyplot as plt
import numpy as np



x = np.linspace(0,10, 100)
y = np.sin(x)
z = np.cos(x)
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
# Sine wave
plt.plot(x,y)
plt.show()
```

```python
# Cos Wave
plt.plot(x,z)
plt.show()
```

```python
plt.plot(y,z)
plt.show()
```

```python
# Adding Title, x-axis, y-axis labels
plt.plot(x,y)
plt.xlabel("angle")
plt.ylabel("sine value")
plt.title("Sine wave")
plt.show()
```

```python
# Parabola
x = np.linspace(-10,10, 20)
y = x**2
plt.plot(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Parabola")
plt.show()
```

```python
# If want the discrete point graph 
plt.plot(x,y, 'gx')   # 'gx' stands for green color and (x symbol)
plt.show()
```

```python
# Multiple Lines
x = np.linspace(-5,5, 100)
plt.plot(x, np.sin(x), 'g')
plt.plot(x, np.cos(x), 'r--')
plt.show()
```

```python
# Bar Graph
f = plt.figure()
ax = f.add_axes([0,0,1,1])
language = ['English', 'French', 'Spanish', 'Latin', 'German']
people = [200, 50, 150, 40, 70]
ax.bar(language, people)
plt.xlabel("Languages")
plt.ylabel("No. of People")
plt.title("Bar Graph")
plt.show()
```

```python
# Pie Chart
f1 = plt.figure()
ax = f1.add_axes([0,0,1,1])
language = ['English', 'French', 'Spanish', 'Latin', 'German']
people = [200, 50, 120, 30, 70]
ax.pie(people, labels = language, autopct = '%1.1f%%') # autopct = auto-pecentage (% -> Opening, 1 -> Padding, .1 -> number of decimals, f -> float datatype, % -> add Percentage symbol, % -> closing)
plt.title("Piechart")
plt.show()
```

```python
# Scatter Plot
x = np.linspace(0,10, 30)
y = np.sin(x)
z = np.cos(x)
a = plt.figure()
ax = a.add_axes([0,0,1,1])
ax.scatter(x,y, color = 'r')
ax.scatter(x,z, color = 'y')
plt.title("Scatter Plot")
plt.show()
```

```python
# 3D Plot
fig = plt.figure()
ax = plt.axes(projection = "3d")
a = np.random.random(200) * 20
x = np.sin(a)
y = np.cos(a)
ax.scatter(x,y,a, c = x, cmap = "Blues")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("3D Plot")
plt.show()
```
